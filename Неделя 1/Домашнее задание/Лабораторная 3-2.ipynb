# Задание 1

Необходимо вывести названия треков, которые были приобретены компаниями, которые обслуживаются сотрудниками, нанятыми после 1 сентября 2002 года. 
    

```python
%%sql
select t.Name, HireDate from 
tracks t
join invoice_items ii on t.TrackId = ii.TrackId
join invoices i on i.InvoiceId = ii.InvoiceId
join customers c on c.CustomerId = i.CustomerId
join employees e on e.EmployeeId = c.SupportRepId
where e.HireDate > '2002-09-01'
```

## Задание 2
Вывести общую стоимость и продолжительность каждого плейлиста.

```python
%%sql
select p.Name as "Playlist", sum(UnitPrice) as "Cost", sum(Milliseconds) as "Length" from
playlists p
join playlist_track pt on p.PlaylistId = pt.PlaylistId
left join tracks t on pt.TrackId = t.TrackId
group by p.Name
```

## Задание 3
Для каждого артиста определить все жанры песен, которые у него есть и количество песен по каждому жанру.


```python
%%sql
select a.Name, g.Name, count(t.Name) from
artists a
join albums al on a.ArtistId = al.ArtistId
join tracks t on al.AlbumId = t.AlbumId
join genres g on g.GenreId = t.GenreId
group by a.name, g.name, a.ArtistId
```

## Задание 4
Для каждого артиста найти жанр, по которому он заработал наибольшее количество денег, а также вывести соотношение общей суммы, заработанной артистом по всем другим жанрам к общей сумме заработанных денег.

```python
%%sql 
with t1 as (select a.ArtistId as "ArtistId",a.Name as "Artist",
            g.Name as "Genre" , sum(t.UnitPrice * ii.Quantity) as "price" 
            from tracks t
            join invoice_items ii on t.TrackId = ii.TrackId
            join albums al on al.AlbumId = t.AlbumId
            join artists a on a.ArtistId = al.ArtistId
            join genres g on g.GenreId = t.GenreId
            group by a.ArtistId, a.Name, g.GenreId, g.Name), 
t2 as (select ArtistId, max(price) as "maxprice",
       sum(price) as "sumprice"
       from t1 
       group by ArtistId),
t3 as (select t1.ArtistId as "ArtistId",t1.Artist as "Artist",
       t1.Genre as "Genre", t1.price as "price", sumprice 
       from t1 join t2 on t1.ArtistId = t2.ArtistId
       where price = maxprice)
select t3.Artist, t3.Genre, price / sumprice from t3
order by t3.Artist
```

## Задание 5
Для каждой компании вывести общее количестов песен, купленных по жанру рок

```python
%%sql
select  company, sum(quantity) as tracks
from tracks t
join genres g on g.genreid=t.genreid
join invoice_items inv_i on inv_i.trackid=t.trackid
join invoices inv on inv.invoiceid=inv_i.invoiceid
join customers c on c.customerid=inv.customerid
where g.name='Rock'
group by company
```

## Задание 6
Вывести альбом, который является минимальным по объему(в байтах) среди альбомов со стоимостью, большей средней цены за альбом по всем альбомам с таким же жанром (если у альбома песни с нескольким жанрами, то жанром является тот, у которого наибольшее количество песен в альбоме)

```python
%%sql 
with T1 (albumid, title, genre, count, bytes, price) as (select al.albumid, title, g.name, count(),sum(bytes),
                                                         sum(t.unitprice)
              from tracks t
              join albums al on al.albumid=t.albumid
              join genres g on g.genreid=t.genreid
             group by al.albumid, al.title, g.genreid, g.name) 
, T2(albumid, title, maxcount) as (select albumid, title, max(count)
        from T1
        group by T1.albumid, T1.title)
, T2N as (select T2.albumid, T2.title, T1.genre, maxcount
         from T1
         join T2 on T2.albumid=T1.albumid
         where T1.count=T2.maxcount) --вывели тут максимум по жанру
, T3(albumid, title, sumbytes, sumprice) as (select albumid, title, sum(bytes), sum(price)
        from T1 
        group by T1.albumid, T1.title) -- вывели объем альбома и стоимость каждого
, T4(genre, avgprice) as (select genre, avg(price)
        from T1
        group by genre)
select T2N.albumid, T2N.title, sumbytes, sumprice, T4.genre, avgprice, min(sumbytes) as bytes
from T3 
join T2N on T2N.albumid=T3.albumid
join T4 on T2N.genre=T4.genre
where sumprice>avgprice
```
